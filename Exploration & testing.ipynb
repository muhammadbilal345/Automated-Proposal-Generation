```python
import json

def generate_proposal(api_input):
    title_prompt = f"Title of the project: {api_input['description']}\n"
    summary_prompt = f"Job Description Summary: {api_input['description']} for {', '.join(api_input['services'])}\n"
    feasibility_prompt = "Technical Feasibility:\n"
    for feature in api_input['features']:
        feasibility_prompt += f"- {feature}\n"
    scope_prompt = "Scope (Milestones, Sprints, Modules {Slight Description}, WorkHours {for each sprint}):\n"
    resources_prompt = "Resources Required (Project Manager, DevOps/Cloud Engineer):\n"
    message_prompt = "This is the AI generated proposal. For specific details and timelines, our team will shortly contact you. (Paraphrased using OpenAI)\n"
    contact_prompt = "Contact Details (Ours):\n"
    
    title = openai.Completion.create(
        engine="text-davinci-003",
        prompt=title_prompt,
        max_tokens=60,
        n=1,
        stop=None,
        temperature=0.7
    ).choices[0].text.strip()
    
    summary = openai.Completion.create(
        engine="text-davinci-003",
        prompt=summary_prompt,
        max_tokens=100,
        n=1,
        stop=None,
        temperature=0.7
    ).choices[0].text.strip()
    
    feasibility = openai.Completion.create(
        engine="text-davinci-003",
        prompt=feasibility_prompt,
        max_tokens=300,
        n=1,
        stop=None,
        temperature=0.7
    ).choices[0].text.strip()
    
    scope = openai.Completion.create(
        engine="text-davinci-003",
        prompt=scope_prompt,
        max_tokens=1000,
        n=1,
        stop=None,
        temperature=0.7
    ).choices[0].text.strip()
    
    resources = [openai.Completion.create(
        engine="text-davinci-003",
        prompt=resources_prompt,
        max_tokens=100,
        n=1,
        stop=None,
        temperature=0.7
    ).choices[0].text.strip() for _ in range(3)]
    
    message = openai.Completion.create(
        engine="text-davinci-003",
        prompt=message_prompt,
        max_tokens=150,
        n=1,
        stop=None,
        temperature=0.7
    ).choices[0].text.strip()
    
    contact = openai.Completion.create(
        engine="text-davinci-003",
        prompt=contact_prompt,
        max_tokens=100,
        n=1,
        stop=None,
        temperature=0.7
    ).choices[0].text.strip()
    
    api_output = {
        "Title": title,
        "Job Description Summary": summary,
        "Technical Fesibility": feasibility,
        "Scope (Milestones, Sprints, Modules {Slight Description}, WorkHours {for each sprint} )": scope,
        "Resources Required (Project Manager, DevOps/Cloud Engineer)": resources,
        "Message": message,
        "Contact Details (ours)": contact
    }
    
    return api_output
```
